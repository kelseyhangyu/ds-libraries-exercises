# Use seaborn's load_dataset function to load the iris database to answer the following questions:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
iris = sns.load_dataset('iris')
iris.head(2)
iris.info()
```

```python
#1 What does the distribution of petal lengths look like?
length = sns.boxplot(iris.petal_length)
```

```python
#2 Is there a relationship between petal length and petal width?
sns.scatterplot(data = iris, x = iris.petal_length,y = iris.petal_width)
```

```python
#3 Would it be reasonable to predict species based on sepal width and sepal length? 
# For this, you'll visualize two numeric columns through the lense of a categorical column.
sns.scatterplot(data = iris, x = iris.sepal_length,y = iris.sepal_width,hue = 'species')
```

```python
#4 Which features would be best used to predict species?
sns.pairplot(iris, hue = 'species', corner = True)
plt.show()
```

```python
sns.lineplot(data = iris, x=iris.species,y=iris.petal_length)
sns.lineplot(data = iris, x=iris.species,y=iris.petal_width)
sns.lineplot(data = iris, x=iris.species,y=iris.sepal_length)
sns.lineplot(data = iris, x=iris.species,y=iris.sepal_width)
```

# Load the chipotle dataset from SQL

```python
# create a bar chart that shows the 4 most popular items and the revenue produced by each.
import env
url =  f'mysql+pymysql://{env.user}:{env.password}@{env.host}/chipotle'

c = pd.read_sql(
    '''
    select * 
from orders
    
    '''
,url)
```

```python
c.head(2)
c.item_price.str.strip('$').astype(float)
c.info()
```

```python
c.head(2)
```

```python
pop = c.groupby('item_name').sum()[['quantity','item_price']].sort_values(by = 'quantity', ascending = False).head(4)
```

```python
pop=pop.sort_values(by = 'item_price', ascending = False)
```

```python
sns.barplot(data=pop, x=pop.index, y=pop.item_price)
```
